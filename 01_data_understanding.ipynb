### Boxplot exploration to identify distribution and outlier 

We noticed that there were alot of extreme outliers, therefore the boxplots didn't provide much information.
In order to extract some information from boxplots we considered all data under the 90th percentile (and > 0 for national_inv)

```python
df_national_inv_90 = df[(df.national_inv > 0) & (df.national_inv < df.national_inv.quantile(.90))]
df_lead_time_90 = df[df.lead_time < df.lead_time.quantile(.90)]
df_in_transit_qty_90 = df[df.in_transit_qty < df.in_transit_qty.quantile(.90)]
df_forecast_3_month_90 = df[df.forecast_3_month < df.forecast_3_month.quantile(.90)]
df_forecast_6_month_90 = df[df.forecast_6_month < df.forecast_6_month.quantile(.90)]
df_forecast_9_month_90 = df[df.forecast_9_month < df.forecast_9_month.quantile(.90)]
df_sales_1_month_90 = df[df.sales_1_month < df.sales_1_month.quantile(.90)]
df_sales_3_month_90 = df[df.sales_3_month < df.sales_3_month.quantile(.90)]
df_sales_6_month_90 = df[df.sales_6_month < df.sales_6_month.quantile(.90)]
df_sales_9_month_90 = df[df.sales_9_month < df.sales_9_month.quantile(.90)]
df_min_bank_90 = df[df.min_bank < df.min_bank.quantile(.90)]
df_potential_issue_90 = df[df.potential_issue < df.potential_issue.quantile(.90)]
df_pieces_past_due_90 = df[df.pieces_past_due < df.pieces_past_due.quantile(.90)]
df_perf_6_month_avg_90  = df[df.perf_6_month_avg < df.perf_6_month_avg.quantile(.90)]
df_perf_12_month_avg_90  = df[df.perf_12_month_avg < df.perf_12_month_avg.quantile(.90)]
df_local_bo_qty_90 = df[df.local_bo_qty < df.local_bo_qty.quantile(.90)]
df_deck_risk_90 = df[df.deck_risk < df.deck_risk.quantile(.90)]
df_oe_constraint_90 = df[df.oe_constraint < df.oe_constraint.quantile(.90)]
df_ppap_risk_90 = df[df.ppap_risk < df.ppap_risk.quantile(.90)]
df_stop_auto_buy_90  = df[df.stop_auto_buy < df.stop_auto_buy.quantile(.90)]
df_rev_stop_90  = df[df.rev_stop < df.rev_stop.quantile(.90)]
```
