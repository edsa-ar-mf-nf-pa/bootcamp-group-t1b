## Import datasets processed in previous step

```python
df=pd.read_hdf(config.df_cleaned, 'df')
df_test=pd.read_hdf(config.df_test_cleaned, 'df')
```

## Add selected new features

```python
def add_new_features(d: pd.DataFrame):
    
    # Add columns that identify perf missing values
    d['perf_6_month_is_na'] = np.where(d.perf_6_month_avg.isna(),1,0)
    d['perf_12_month_is_na'] = np.where(d.perf_12_month_avg.isna(),1,0)
    d['lead_time_is_na'] = np.where(d.lead_time.isna(),1,0)

    # Add column that identify negative inventory values
    d['national_inv_neg'] = np.where(d.national_inv < 0, 1, 0)

    d['forecast_evol_36'] = d.apply(lambda x: fn.variation(x['forecast_3_month']/3, x['forecast_6_month']/6), axis=1)
    d['forecast_evol_39'] = d.apply(lambda x: fn.variation(x['forecast_3_month']/3, x['forecast_9_month']/9), axis=1)
    d['forecast_evol_69'] = d.apply(lambda x: fn.variation(x['forecast_6_month']/6, x['forecast_9_month']/9), axis=1)

    d['sales_evol_13'] = d.apply(lambda x: fn.variation(x['sales_1_month'], x['sales_3_month']/3), axis=1)
    d['sales_evol_16'] = d.apply(lambda x: fn.variation(x['sales_1_month'], x['sales_6_month']/6), axis=1)
    d['sales_evol_19'] = d.apply(lambda x: fn.variation(x['sales_1_month'], x['sales_9_month']/9), axis=1)
    d['sales_evol_36'] = d.apply(lambda x: fn.variation(x['sales_3_month']/3, x['sales_6_month']/6), axis=1)
    d['sales_evol_39'] = d.apply(lambda x: fn.variation(x['sales_3_month']/3, x['sales_9_month']/9), axis=1)
    d['sales_evol_69'] = d.apply(lambda x: fn.variation(x['sales_6_month']/6, x['sales_9_month']/9), axis=1)
```

```python
add_new_features(df)
add_new_features(df_test)
```
